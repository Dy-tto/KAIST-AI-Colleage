### 1) Missing Data 처리하기
#### np.nan
- Missing value는 Nan으로 나타나며, np.nan으로 지정할 수 있다.

```python
df
```

```python
df['I']=[1,np.nan,2,np.nan,3,True,np.nan]
df
```

```python
# pandas 1.1.5
print(pd.isna(df['I'][1]))
print(pd.isnull(df['I'][1]))
print(np.isnan(df['I'][1]))
```

```python
a = np.nan
b=np.nan
a==b
```

```python
type(np.nan)
```

```python
df
```

```python
df.loc['2020-12-06',:]=[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
df
```

#### dropna()
* Nan값을 없애는 데에 쓰일 수 있는 함수
1. how=’any’ : 행의 값 중 하나라도 nan일 경우 그 행을 없앤다.
2. how=‘all’ : 행의 값이 모두 nan일 경우 그 행을 없앤다.

```python
dropped_df = df.dropna(how='any')
dropped_df
```

```python
dropped_df = df.dropna(how='all')
dropped_df
```

#### fillna()
Nan값에 다른 값을 넣는 함수

```python
df
```

```python
filled_df = df.fillna(value='9.9')
filled_df
```

#### isnull()
Nan인지 확인하는 함수

```python
check_df = df.isnull()
check_df
```

```python
pd.isnull(df)
```

이를 이용하여 특정 열에서 nan값을 가지는 행만을 추출하기

```python
df.loc[df.isnull()['I'],:]
```
