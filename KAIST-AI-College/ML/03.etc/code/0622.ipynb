```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(iris.data, iris.target ,  test_size = 0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Classification > 나이브 베이즈 

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
import graphviz
from sklearn.tree import export_graphviz
```

```python
dot_data = export_graphviz(dt,out_file = None, feature_names = iris.feature_names,
                           class_names = iris.target_names)
```

```python
graph = graphviz.Source(dot_data)
```

```python
graph
```

# k-Nearest Neighbors 

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors = 5, metric='minkowski',p=2)
```

```python
neigh.fit(X_train,y_train)
```

```python
y_pred = neigh.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

# SVM


```python
from sklearn.svm import SVC
svc = SVC(kernel = "linear")
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
x_min, x_max = iris.data[:, 0 ].min() -1, iris.data[:,0].max() +1
y_min, y_max = iris.data[:, 1 ].min()  -1, iris.data[:,1].max() +1
```

```python
iris.data[:,0]
```

```python
iris.data[:, 1 ]
```

```python
svc_linear = SVC(kernel = 'linear').fit(iris.data[:,0:2],iris.target)
svc_poly = SVC(kernel = 'poly').fit(iris.data[:,0:2],iris.target)
svc_rbf = SVC(kernel = 'rbf').fit(iris.data[:,0:2],iris.target)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def vis_svc(svc):
    # Return coordinate matrices from coordinate vectors
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(5, 5))
    # plot contour
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm)
    # plot training points
    plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap=plt.cm.coolwarm)
    plt.xlabel('sepal length')
    plt.ylabel('sepal width')
```

```python
vis_svc(svc_linear)
```

```python
vis_svc(svc_poly)
```

```python
vis_svc(svc_rbf)
```
